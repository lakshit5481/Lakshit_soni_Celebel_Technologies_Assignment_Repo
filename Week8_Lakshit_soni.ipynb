# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION
import re

def agent(query: str):
    """
    A simple rule-based smart assistant that routes user queries
    to the appropriate tool and returns a structured response.
    """
    try:
        query_lower = query.lower().strip()

        # Routing  calculation requests
        if "calculate" in query_lower:
            expression = query_lower.replace("calculate", "").strip()
            return {
                "type": "calculation",
                "result": calculator(expression)
            }

        # Routign keyword extraction requests
        elif "keyword" in query_lower:
            text = re.sub(r"extract\s+keywords\s+from", "", query, flags=re.IGNORECASE).strip()
            return {
                "type": "keywords",
                "result": extract_keywords(text)
            }

        # Handle general questions
        else:
            import google.generativeai as genai

            genai.configure(api_key=API_KEY) # please enter your api key!!

            model = genai.GenerativeModel("gemini-2.5-flash")

            response = model.generate_content(query)

            return {
                "type": "general",
                "response": response.text
            }

    except Exception as e:
        return {
            "type": "error",
            "result": f"Something went wrong: {str(e)}"
        }
````
